## List Modifications (Easy)

```python
# get number of operations to be performed from user
N = int(input())

# create the empty list
output = []

# create a dictionary of all possible operations and the associated lamda functions
operations = {
    "insert" : lambda x, y, z: x.insert(y, z),
    "print" : lambda x: print(x),
    "remove" : lambda x, y: x.remove(y),
    "append" : lambda x, y: x.append(y),
    "sort" : lambda x: x.sort(),
    "pop" : lambda x: x.pop(),
    "reverse" : lambda x: x.reverse()
}

# iterate through commands and peform the operations on the list
for i in range(N):
    # split input at each space into list of values
    op = input().split()
    command = op[0]
    try:
        operations[command](output, int(op[1]), int(op[2]))
    except IndexError:
        try:
            operations[command](output, int(op[1]))
        except IndexError:
            operations[command](output)
#     print(output)
```

```python
# alternate solution
def execute(lst, cmd, *args):
    if cmd == 'insert':
        lst.insert(int(args[0]), int(args[1]))
    elif cmd == 'print':
        print(lst)
    elif cmd == 'remove':
        lst.remove(int(args[0]))
    elif cmd == 'append':
        lst.append(int(args[0]))
    elif cmd == 'sort':
        lst.sort()
    elif cmd == 'reverse':
        lst.reverse()
    elif cmd == 'pop':
        lst.pop()
    else: 
        print("Command not recognized!")

lst = []
for _ in range(int(input())):
    execute(lst, *input().split())
```

## Mutations (Easy)
Write a function to change one character in a string at a specific index

```python
def mutate_string(string, position, character):
    l = list(string)
    l[position] = character
    string = ''.join(l)
    return string
```

```python
mutate_string('adamtalbot', 4, 'a')
```

```python
def mutate_string2(string, position, character):
    string = string[:position] + character + string[position+1:]
    return string
```

```python
mutate_string2('adamtalbot', 4, 'a')
```

## Find a String (Easy)
In this challenge, the user enters a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left.

NOTE: String letters are case-sensitive.

use while loop to keep checking if substring is there and then slice each time one ahead of index, have counter to count instances

```python
s1 = 'ABCDCDC'

s2 = 'CDC'

r = 0
counter = -1
while r != -1:
#     print(s1)
    r = s1.find(s2)
    print(r)
    s1 = s1[r+1:]
    counter += 1
#     print(counter)
print(counter)
```

```python
def count_substring(string, sub_string):
    r = 0
    counter = -1
    while r != -1:
        r = string.find(sub_string)
        string = string[r+1:]
        counter += 1
    return counter
```

```python
# another very similar solution
def count_substring(string, sub_string):
    count = 0
    i = string.find(sub_string)
    while i != -1:
        count += 1
        i = string.find(sub_string, i+1)
    return count
```

## String Validators (Easy)
You are given a string .
Your task is to find out if the string  contains: alphanumeric characters, alphabetical characters, digits, lowercase and uppercase characters.

```python
s = 'qA2'

# create a list of booleans for each char in the string for each method
# then check if the list contains any true values
alnumList = []
alphaList = []
digitList = []
lowerList = []
upperList = []

for char in s:
    alnumList.append(char.isalnum())
    alphaList.append(char.isalpha())
    digitList.append(char.isdigit())
    lowerList.append(char.islower())
    upperList.append(char.isupper())

print(True in alnumList)
print(True in alphaList)
print(True in digitList)
print(True in lowerList)
print(True in upperList)
```

```python
# cleaner solution

s = 'qA2'
print(any(c.isalnum() for c in s))
print(any(c.isalpha() for c in s))
print(any(c.isdigit() for c in s))
print(any(c.islower() for c in s))
print(any(c.isupper() for c in s))
```

## Set .discard(), .remove() & .pop() (Easy)
You have a non-empty set , and you have to execute  commands given in  lines.

The commands will be pop, remove and discard.

Input Format

The first line contains integer , the number of elements in the set .
The second line contains  space separated elements of set . All of the elements are non-negative integers, less than or equal to 9.
The third line contains integer , the number of commands.
The next  lines contains either pop, remove and/or discard commands followed by their associated value.

Output Format

Print the sum of the elements of set  on a single line.

```python
# get number of entries in set
n = int(input())

# get the set to start with
s = set(map(int, input().split()))

# get number of operations to be performed from user
N = int(input())

# create a dictionary of all possible operations and the associated lamda functions
operations = {
    "discard" : lambda x, y: x.discard(y),
    "remove" : lambda x, y: x.remove(y),
    "pop" : lambda x: x.pop()
}

# iterate through commands and peform the operations on the list
for i in range(N):
    # split input at each space into list of values
    op = input().split()
    command = op[0]
    try:
        operations[command](s, int(op[1]))
    except IndexError:
        try:
            operations[command](s)
        except IndexError:
            pass
print(sum(s))
```

```python
# try solution with if and elifs
# get number of entries in set
n = int(input())

# get the set to start with
s = set(map(int, input().split()))

# get number of operations to be performed from user
N = int(input())

for i in range(N):
    # split input at each space into list of values
    op = input().split()
    command = op[0]
    if command == 'pop':
        s.pop()
        print(s)
    elif command == 'remove':
        s.remove(int(op[1]))
        print(s)
    elif command == 'discard':
        s.discard(int(op[1]))
        print(s)
    else:
        print('unrecognized command')
print(sum(s))
```

## collections.Counter()
 is a shoe shop owner. His shop has  number of shoes.
He has a list containing the size of each shoe he has in his shop.
There are  number of customers who are willing to pay  amount of money only if they get the shoe of their desired size.

Your task is to compute how much money  earned.

```python
shoe_count = int(input())
shoe_sizes = [int(x) for x in input().split()]
num_customers = int(input())
profit = 0
for i in range(num_customers):
    size, price = [int(x) for x in input().split()]
    if size in shoe_sizes:
        profit += price
        shoe_sizes.remove(size)
        print(shoe_sizes)
        print(profit)
print(profit)
```
